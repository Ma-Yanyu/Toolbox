```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Read the CSV file into a DataFrame
df = pd.read_csv('/dataStor/home/yyma/KGC-project/CM-BKG/src/graph_partition/random_walk_stat.csv')  # Replace 'your_file.csv' with the actual file path

# Step 2: Set up the figure and axes
fig, ax1 = plt.subplots(figsize=(12, 8))

# Plot the first set of metrics on the primary y-axis
ax1.set_xlabel('Start Size (Number of Nodes for Random Walk)', fontsize=12)
ax1.set_ylabel('Entropy and Avg Degrees', fontsize=12)

# Plot entropy
ln1 = ax1.plot(df['start_size'], df['entropy'], marker='o', color='b', label='Entropy')

# Plot avg_in_degree and avg_out_degree on the same primary y-axis
# ln2 = ax1.plot(df['start_size'], df['avg_in_degree'], marker='o', color='g', label='Avg In-Degree')
ln3 = ax1.plot(df['start_size'], df['avg_out_degree'], marker='o', color='r', label='Avg Out-Degree')

# Create a secondary y-axis to plot metrics with a different scale
ax2 = ax1.twinx()  # Instantiate a second y-axis sharing the same x-axis

# Plot density on the secondary y-axis
ax2.set_ylabel('Density, Num Nodes, Num Edges', fontsize=12)
ln4 = ax2.plot(df['start_size'], df['density'], marker='o', color='c', label='Density')

# Combine all labels into a single legend
lns = ln1 + ln3 + ln4
labels = [l.get_label() for l in lns]
ax1.legend(lns, labels, loc='best')

# Customize the plot
plt.title('Graph Metrics vs Start Size', fontsize=16)
plt.grid(True)

# Show the plot
plt.show()
```
